# NN for classification in-out passengers

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
import gdown, os, random

%matplotlib inline
```

```python
gdown.download('https://storage.googleapis.com/terra_ai/DataSets/bus.zip', None, quiet = True)
```

```python
base_dir = 'bus'
x_train = []
y_train = []
img_height = 128
img_width = 64
```

```python
for in_out in os.listdir(base_dir):
    for img in os.listdir(base_dir + '/' + in_out):
        x_train.append(image.img_to_array(image.load_img(base_dir + '/' + in_out + '/' + img,
                                                        target_size = (img_height, img_width))))
        if in_out == 'entering':
            y_train.append(0)
        else:
            y_train.append(1)
```

```python
x_train = np.array(x_train)
y_train = np.array(y_train)

print(x_train.shape, y_train.shape)
```

```python
x_train_reshape = x_train.reshape(-1, img_widthidth * img_height * 3)
print(x_train_reshape.shape)
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_train_reshape, 
                                                  y_train, 
                                                  test_size=0.2, 
                                                  shuffle=True, 
                                                  random_state=3)

print('Размеры сформированных массивов:')

print(x_train.shape)
print(x_val.shape)
print(y_train.shape)
print(y_val.shape)
```

```python
model = Sequential()
model.add(Dense(2048, input_dim = x_train.shape[1], activation = 'relu'))
model.add(Dense(1024, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = Adam(learning_rate = 0.00001), metrics = ['accuracy'])
```

```python
history = model.fit(x_train, y_train, epochs = 25, verbose = 1, batch_size = 64, validation_split = 0.2, shuffle = True)
```

```python
plt.plot(history.history['val_loss'])
plt.plot(history.history['val_accuracy'])
plt.show()
```
